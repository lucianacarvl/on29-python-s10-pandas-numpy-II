Importando os dados do arquivo csv

```python
import pandas as pd
import numpy as np
df = pd.read_csv('data_sample_.csv', sep=';' , encoding='latin-1', on_bad_lines='skip')
```

```python
df
```

Criando os dados para o gráfico de linhas

Tratar a hora para o formato XX:XX
Agrupar pela Hora e fazer a média da temperatura

```python
df['Hora'] = df['Hora UTC'].str.replace('00 UTC',':00')
```

```python
temperatura_hora = df.groupby(['Hora'])['TEMPERATURA DO AR - BULBO SECO. HORARIA (C)'].mean().to_numpy()
```

```python
vento_hora = df.groupby(['Hora'])['VENTO. VELOCIDADE HORARIA (m/s)'].mean().to_numpy()
```

```python
pressao_hora = df.groupby(['Hora'])['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO. HORARIA (mB)'].mean().to_numpy()
```

```python
sensacao_hora = 33 + (10 * np.sqrt(vento_hora) + 10.45 - vento_hora) * ((temperatura_hora - 33)) / 22
```

```python
horas = np.arange(24)
```

## Colocar uma descrição do eixo X

Para colocar uma descrição no eixo X use a função `plt.xlabel()`, passando como parametro o texto do label

## Colocar uma descrição do eixo Y

Para colocar uma descrição no eixo Y use a função `plt.ylabel()`, passando como parametro o texto do label

## Titulo do gráfico

Para colocar o título no gráfico use a função `plt.title()`, passando como parametro o texto do título

```python
plt.xlabel("Horas")
plt.ylabel("Temperatura")
plt.title("Temperatura por hora")

plt.plot(temperatura_hora, color = 'r')
plt.plot(sensacao_hora, color = 'g')
```

## Fonte dos Labels e titulo

Para alterar a font use o parametro `fontdict` nas funções `xlabel()`, `ylabel()`, and `title()`

Ele usa um objeto com os seguintes propriedades:
*   family
*   color
*   size

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Horas",fontdict = font2)
plt.ylabel("Temperatura",fontdict = font2)
plt.title("Temperatura por hora",fontdict = font1)

plt.plot(temperatura_hora, color = 'r')
plt.plot(sensacao_hora, color = 'g')
```

## Habilitando as linhas de referencia

para habilitar as linhas chame a função `plt.grid()`

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Horas",fontdict = font2)
plt.ylabel("Temperatura",fontdict = font2)
plt.title("Temperatura por hora",fontdict = font1)

plt.grid()

plt.plot(temperatura_hora, color = 'r')
plt.plot(sensacao_hora, color = 'g')
```

## Linhas de referencia
é possivel abilitar as linhas de referencia em determinado sentido, para isso usamos o parametro `axis`, com os seguintes dados

|Atributo|Eixo|
|--|--|
|'x'|horizontal|
|'y'|vertical|
|'both'|ambos|

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Horas",fontdict = font2)
plt.ylabel("Temperatura",fontdict = font2)
plt.title("Temperatura por hora",fontdict = font1)

plt.grid(axis = 'x')

plt.plot(temperatura_hora, color = 'r')
plt.plot(sensacao_hora, color = 'g')
```

## Estilo das linhas guias

assim como as linhas do gráfico podemos alterar o estilo da linha com os parametros `color`, `linestyle` e `linewidth`, é possivel consultar como colocar os valores acima

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Horas",fontdict = font2)
plt.ylabel("Temperatura",fontdict = font2)
plt.title("Temperatura por hora",fontdict = font1)

plt.grid(color = 'b', linestyle = '--', linewidth = 0.5)

plt.plot(temperatura_hora, color = 'r')
plt.plot(sensacao_hora, color = 'g')
```

# Gráfico de Disperssão
Imprimendo um gráfico de dispersão use a função `plt.scatter()`, passando como primeiro parametro os valores do eixo X e segundo parametro os valores do eixo Y

```python
plt.scatter(vento_hora, temperatura_hora)
```

## Mais um conjunto de pontos
Para mais um conjunto de pontos chame outra vez a função `plt.scatter()`, o único porem é que o eixo X deve ser o mesmo para as duas funções

```python
plt.scatter(vento_hora, temperatura_hora)

plt.scatter(vento_hora, sensacao_hora)
```

## Cores dos pontos
Para mudar a cor dos pontos podemos usar o parametro `colors` ou sua abreviação `c`, e as cores seguem o mesmo padrão do gráfico de linhas

```python
plt.scatter(vento_hora, temperatura_hora, color = '#88c999')
```

## Variando a cor dos pontos
As cores dos pontos podem variar com uma tabela de numeros, assim adicionamos mais informação para o gráfico de dispersão

```python
plt.scatter(vento_hora, temperatura_hora, c= vento_hora)
```

## Tamanho do ponto
O tamanho do ponto pode ser alterado usando o parametro `s`, ele pode receber um numero inteiro ou pode assim como a cor receber um array de informação tb

```python
plt.scatter(pressao_hora, temperatura_hora, s = (vento_hora*100))
```

## Transparencia dos pontos
para mexer na transparencia dos dados use o parametro `alpha`, ele varia de 0 ~ 1



```python
plt.scatter(pressao_hora, temperatura_hora, s = (vento_hora*100), alpha=0.5)
```

```python
plt.scatter(pressao_hora, temperatura_hora, s = (vento_hora*100), alpha=0.5, c= vento_hora)
```

# Gráfico de Barras
Imprimendo um gráfico de barras use a função `plt.bar()`, passando como primeiro parametro os valores do eixo X e segundo parametro os valores do eixo Y

```python
plt.bar(horas, temperatura_hora)
```

## Mudando cor da coluna
Para mudar a cor da coluna use o parametro `color`, as cores seguem o padrão que nem o gráfico de linhas

```python
plt.bar(horas, temperatura_hora, color='r')
```

## Mudando o tamanho da coluna
Para mudar o tamanho da coluna user o parametro `width`

```python
plt.bar(horas, temperatura_hora, color='r', width = 0.1)
```

## Gráfico de Colunas
Segue os mesmos padrões do gráfico de baras, e os mesmos parametros, só o parametro `width` que é `height` nesse caso, a unica coisa que muda é a função agora chama `plt.barh()`

```python
plt.barh(horas, temperatura_hora, color='r', height = 0.1)
```
